```python
def G(gamma):
    A = np.array([-0.577191652, 0.988205891, -0.897056937, 0.918206857,
                  -0.756704078, 0.482199394, -0.193527818, 0.035868343])
    s = 0
    for i in range(8):
        s = s + A[i] * (gamma-1)**(i+1)
    return 1 + s
```

```python
# The distribution function of C7+ in term of Molecular weight
for gamma in [1 , 1.2, 2]:
    Md = 165
    tau = 86
    Beta = (Md - tau)/gamma
    # C6 to C45
    M = np.array([84, 96, 107, 121, 134, 147, 161, 175, 190, 206, 222, 237, 251, 263, 275, 291, 300, 312, 324, 337, 349, 360, 372, 382, 394, 404, 415, 426, 437, 445, 456, 464, 475, 484, 495, 502, 512, 521, 531, 539])
    #F = 0.0376*np.exp(-M/Beta)
    F = ( (M-tau)**(gamma-1) * np.exp(-(M-tau)/Beta) ) / (Beta**gamma *G(gamma))

    plt.plot(M, F)
    plt.title('Gamma probability function')
    plt.ylabel('F(M)')
    plt.xlabel('Molecular Weight')
```

```python
def F_cal(M):
    gamma = 1
    Md = 165
    tau = 86 # 86 and 90
    Beta = (Md - tau)/gamma
    F = ( (M-tau)**(gamma-1) * np.exp(-(M-tau)/Beta) ) / (Beta**gamma *G(gamma))
    return F
```

```python
split = 45
Md = 165
tau = 90
gamma = 1
Beta = (Md - tau)/gamma
Mn = np.array([90 , 101.5, 114 , 127.5, 140.5, 154 , 168, 182.5, 198, 214, 229.5, 244, 257, 269, 283, 295.5, 306, 318, 330.5, 343, 354.5, 366, 377, 388, 399, 409.5, 420.5, 431.5, 441, 450.5, 460, 469.5, 479.5, 489.5, 498.5, 507, 516.5, 526, 535, 544])
print('Beta:', Beta)
```

```python
Zcn = np.zeros(split-7+1)
for i in range(split-7+1):
    Mn_plus = Mn[i+1]
    Mn_minus = Mn[i]
    Zcn[i] = -np.exp(tau/Beta) * (np.exp(-Mn_plus/Beta) - np.exp(-Mn_minus/Beta)) 
    print(f'C{i+7}:',Mn_minus, Mn_plus, Zcn[i])
print('Sum Zcn:', (Zcn * 3.92).sum())
print(Zcn * 3.92)
```
